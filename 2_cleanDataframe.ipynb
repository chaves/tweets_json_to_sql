```python
import pandas as pd
```

```python
OUTPUT = '_output'
```

```python
# Merge volkswagen and wv dataframes
df1 = pd.read_pickle('{}/volkswagen_old_total_brut.pickle'.format(OUTPUT))
df2 = pd.read_pickle('{}/vw_old_total_brut.pickle'.format(OUTPUT))
df = df1.append(df2, ignore_index=True)
df.info()
```

```python
# Tweets with no date
df[df['time']==0]
```

```python
# Drop tweets with no date
df = df.drop(df[df.time ==0].index)
# It worked ?
df[df['time']==0]
```

```python
# Drop duplicates
df = df.drop_duplicates(subset=['id'])
```

```python
df.info()
```

```python
def get_date_splits(date_in):
    dates_out = 0
    if (type(date_in) == str):
        date_in = date_in.split('.000Z')
        dates_out = date_in[0].split('T')
    return dates_out

def get_date(date_in):
    dates_out = get_date_splits(date_in)
    return dates_out[0]

def get_time(date_in):
    dates_out = get_date_splits(date_in)
    return dates_out[1]
```

```python
# Clean date formats
df['date'] = df['time'].apply(lambda x: get_date(x))
df['time_new'] = df['time'].apply(lambda x: get_time(x))
```

```python
# Drop bad date format
df.drop(['time'], axis=1)
df.head(2)
```

```python
# save clean file
df.to_pickle('{}/total_old_clean.pickle'.format(OUTPUT))
```
